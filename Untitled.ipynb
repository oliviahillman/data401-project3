```python
from math import sqrt

size = 0.5

def hex_to_pixel(coord):
    x = (sqrt(3) * coord[0]  +  sqrt(3)/2 * coord[1])
    y = (                         3./2 * coord[1])
    return (x, y)

def hex_to_pixel_2(coord):
    x = size * (     3./2 * coord[0]                      )
    y = size * (sqrt(3)/2 * coord[0]  +  sqrt(3) * coord[1])
    return (x, y)
```
